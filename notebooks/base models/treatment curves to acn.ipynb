# Visualizaci√≥n de curvas y tratamiento para recomendaciones online

```python
import pickle
import pandas as pd
import numpy as np
from scipy.signal import savgol_filter
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
import warnings
import matplotlib.pyplot as plt
from plotly.subplots import make_subplots
warnings.filterwarnings("ignore")
```

```python
# Se carga la curva de celda de carga v/s granulometria (dado un subcontexto)
dic1 =  pickle.load(open('..//..//models//cc_recommendations.pkl', 'rb'))
```

```python
titulos=tuple(dic1.keys())
```

```python
#dic1['23'].set_index('gran').plot()
#plt.show()
```

```python
# Mostrando curvas de recomendaci√≥n
fig = make_subplots(
    rows=4, cols=3,
    subplot_titles=titulos,
    )
k=0
for i in range(1,5):
    for j in range(1,4):
        try:
            data=dic1[titulos[k]]
            fig.add_trace(go.Scatter(x=data["gran"], y=data["cc"], ##FF6511
                            mode='lines',
                            name=titulos[k],line=dict(width=3, ),legendgroup = '1'),row=i, col=j) 

            k+=1
        except:
            pass
     



fig.update_layout(height=800, width=1200, title_text="Curvas de recomendaci√≥n celda de carga vs granulometr√≠a",legend_title="Sub-contexto (Edad-Cluster)"#,xaxis6_rangeslider_visible=True
,#template="plotly_dark"
)

fig.show()
```

```python
# Se genera dataframe con indices que representan intervalos de granulometria
dic2={}

for i in dic1.keys():
    tmp = dic1[i].set_index('gran')
    left=list((pd.DataFrame(tmp.index).shift(1)+0.5).fillna(0)["gran"])
    right=list((pd.DataFrame(tmp.index)+0.5)["gran"])
    right[-1]=100
    tmp.index=pd.IntervalIndex.from_arrays(
                            left, right)
    dic2[i]=tmp
```
