# Causality test

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.stattools import grangercausalitytests
```

```python
# read
tags=pd.read_excel("../../data/meta data/tags relevantes.xlsx")
tags_cc=tags.tag.to_list()
dic={}
for i,j in zip(tags.tag,tags.description):
    dic[i]=j
tags
```

```python
variables_interes=['pyrite law', 'chalcopyrite law',
       'chalcocite law', 'covelin law', 'crusher index', 'sag power index',
       'ball work index', 'bornite law','charge cell', 'TPH','water', 'speed','HH TPH', 'granulometry',
       "power","solid percentage","HH charge cell","LL charge cell"]
variables_interes
```

```python
# Se lee csv
cleaned=pd.read_csv('../../data/consolidated data/cleaned_15December2022.csv', parse_dates=['Timestamp'], index_col='Timestamp').dropna()
cleaned.sort_index(inplace=True)
cleaned.rename(columns=dic,inplace=True)
cleaned=cleaned[variables_interes]
cleaned.head()
```

# Test de causalidad de Wiener-Granger

```python
data =cleaned[['sag power index', 'LL charge cell']].dropna()
grangercausalitytests(data, maxlag=3)
```

```python
data =cleaned[['TPH','speed']].dropna()
grangercausalitytests(data, maxlag=3)
```

```python
data =cleaned[['solid percentage','water']].dropna()
grangercausalitytests(data, maxlag=3)
```
